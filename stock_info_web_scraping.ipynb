```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
sp500_df = pd.read_csv("sp500_05_17_2023.csv")
sp500_df.head()
```

```python
tickers = sp500_df['Symbol'].tolist()
```

```python
stock_info = pd.DataFrame(columns=['Ticker', 'Name', 'Sector', 'Market Cap', 'Share Price', 'Revenue', 'Net Income', 'Summary'])
stock_info['Ticker'] = tickers
stock_info['Name'] = sp500_df['Name']
stock_info['Share Price'] = sp500_df['Last']
stock_info.head()
```

```python
for i in range(len(tickers)):
    try:
        ticker_info = yf.Ticker(tickers[i]).info
        
        try:
            stock_info['Sector'][i] = ticker_info['sector']
            stock_info['Market Cap'][i] = ticker_info['marketCap']
            stock_info['Revenue'][i] = ticker_info['totalRevenue']
            stock_info['Net Income'][i] = ticker_info['netIncomeToCommon']
            stock_info['Summary'][i] = ticker_info['longBusinessSummary']
        except:
            print(tickers[i], " is missing info on Yahoo Finance")
    except: 
        print(tickers[i], " could not be found on Yahoo Finance")
```

```python
stock_info.dropna()
stock_info.head()
```

```python
stock_info.to_csv("stock_info.csv")
```
